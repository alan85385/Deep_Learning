# Class

```python
class Man:
    
    def __init__(self, name):
        
        self.name = name
        print('Initialized!')
        
    def hello(self):
        
        print('Hello ' + self.name + '!')
        
    def goodbye(self):
        
        print('Good-Bye ' + self.name + '!')
```

```python
m = Man('Alan')
m.hello()
m.goodbye()
```

# Numpy & Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label='sin')
plt.plot(x,y2,linestyle='--',label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('sin & cos')
plt.show()
```
